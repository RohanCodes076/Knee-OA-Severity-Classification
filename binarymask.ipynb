```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Load X-ray image (grayscale)
xray_image = cv2.imread(r'H:\2nd_year_practice\knee_OA_detection_withApp\knee_OA_dl_app-main\KneeOADataset\train\3\9002430L.png', cv2.IMREAD_GRAYSCALE)

# Assuming the ROI is significantly brighter than the background
_, binary_mask = cv2.threshold(xray_image, 150, 255, cv2.THRESH_BINARY)

# Apply the mask as before
masked_image = xray_image * (binary_mask // 255)

# Display the original X-ray, the mask, and the masked image
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(xray_image, cmap='gray')
plt.title('Original X-ray Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_mask, cmap='gray')
plt.title('Binary Mask')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(masked_image, cmap='gray')
plt.title('Masked X-ray Image (ROI)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
